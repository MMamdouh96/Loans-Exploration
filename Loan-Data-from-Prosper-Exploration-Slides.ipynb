## Dataset Overview

> The data contains data about loans it's 113937 rows and 81 columns. we have many important columns (LoanStatus, BorrowerRate, ListingCategory (numeric), IncomeRange, IncomeVerifiable, StatedMonthlyIncome)

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv("prosperLoanData.csv")
```

## Distribution of Loan Original Amount

> As we see the amount of loans starting from less that 250 dollar to 35000 dollar as we see, and we can see that also the most common loan type around 5000 them 15000 then 10000.

```python
plt.figure(figsize=[8, 5])

log_binsize = 0.025
bins = 10 ** np.arange(2.4, np.log10(df['LoanOriginalAmount'].max())+log_binsize, log_binsize)
df['LoanOriginalAmount'].hist(bins=bins);
plt.xlabel('Amount $');
plt.title("Distribution of Loan Original Amount");
```

## Distribution of Stated Monthly Income

> For the Distribution of Stated Monthly Income, we can see that we have found an outlier in the distribution with Stated Monthly Income 1750000 that is really big number comparing to the all data so we have filtered it out, to see the real distribution more clear. Form the distribution of Stated Monthly Income we can notice that the average of monthly income is around 5000 and that which means the people who get paid around 5000 has higher percentage to take loans.

```python
binsize = 100
bins = np.arange(1000, 50000 + binsize, binsize)

plt.figure(figsize=[8, 5]);
df.boxplot(column=['StatedMonthlyIncome']);
plt.title("Boxplot of Stated Monthly Income");
plt.show();

plt.figure(figsize=[8, 5]);
df.query("StatedMonthlyIncome < 150000")['StatedMonthlyIncome'].hist(bins=bins);
plt.xlabel('Amount $');
plt.title("Distribution of Stated Monthly Income");
plt.show()
```

## Relationship between LoanOriginalAmount and Investors

> As we see the amount of investors increases with amount of the loan to specific point that is around 25000 after that we can not find such a pattern but it increases again in loan with 35000 value.

```python
df.plot("LoanOriginalAmount", "Investors", kind='scatter', figsize=[12, 6]);
plt.title("Relationship between LoanOriginalAmount and Investors");
```
